```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from scipy.linalg import svd
from scipy.spatial import distance
import warnings
warnings.filterwarnings('ignore')
```

# Part 1.1 Elbow and K-Means

```python
"""Helper Functions"""

def make_dataset():
    X1, _ = make_blobs(n_samples=50, centers=[[0, 5]], cluster_std=[2.5], random_state=42)
    X2, _ = make_blobs(n_samples=50, centers=[[4, -1]], cluster_std=[1.5], random_state=42)
    X3, _ = make_blobs(n_samples=50, centers=[[-2, 2]], cluster_std=[0.8], random_state=42)
    X4, _ = make_blobs(n_samples=50, centers=[[-5, 0]], cluster_std=[1], random_state=42)
    return np.vstack([X1, X2, X3, X4])

def show_data_results(X, num_plot=2, y_pred=None, cmap='jet'):
    """
    Plots the input data and additionally predicted clusters if provided.
    Set num_plot=1 to plot only input.
    Set num_plot=2 along with a y_pred to plot predicted clusters.
    """
    if num_plot==1:
        plt.scatter(X[:, 0], X[:, 1])
        plt.title ("input data")
    elif num_plot==2:
        try:
            assert y_pred is not None
            fig = plt.figure(figsize=(10, 3))
            ax1 = fig.add_subplot(121)
            ax1.scatter(X[:, 0], X[:, 1])
            ax1.set(xticks=[],yticks=[],title ="input data")

            ax2 = fig.add_subplot(122)
            ax2.scatter(X[:, 0], X[:, 1], c=y_pred, cmap=cmap)
            ax2.set(xticks=[], yticks=[], title ="clustered data")
        except:
            print('y_pred is required for 2 plots')
```

```python
X = make_dataset()

""" TODO: run K-means and plot Elbow"""
obj = []
K = range(1,10)
for k in K:
   kmeanModel = KMeans(n_clusters=k).fit(X)
   kmeanModel.fit(X)
   
   ### TODO: implement objective function calculation here
   obj.append(kmeanModel.inertia_)
```

```python
##Plot the elbow vs K
plt.plot(K, obj, 'bx-')
plt.xlabel('k')
plt.ylabel('Objective Function Value')
plt.title('The Elbow Method showing the optimal k')
plt.show()

""" Uncomment this to check the actual data """
#show_data_results(X, num_plot=1) 
```

From this plot it appears that either 3 or 4 would be the optimal number of clusters.  Based on the visualization of the data in the uncommented code I would say that 4 is the optimal number of clusters.

# Part 1.2(b)

```python
""" 2(b) Adjacency Matrix Implementation 
TODO: implement the adjacency matrix calculation"""
def get_adjacency_matrix(gamma, X):
    adjacency_matrix = None
    ### Your code here ###
    n = X.shape[0]
    adjacency_matrix = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            adjacency_matrix[i,j] = np.exp(-gamma * distance.euclidean(X[i], X[j])**2)

    return adjacency_matrix
```

Increasing gamma would serve to reduce the value of the output.  An increased gamma would have a stronger influence on points that are nearby, but a smaller impact on points that are further away.  A decreased value of gamma would result in a more uniform influence on all points of the adjacency matrix.

# Part 1.2(c)

```python
""" 2(c) Degree Matrix Implementation 
TODO: implement the degree matrix calculation"""
def get_degree_matrix(adjacency_matrix):
    degree_matrix = None
    ### Your code here ###
    degree_matrix = np.diag(np.sum(adjacency_matrix, axis=1))
    
    return degree_matrix
```

# Part 1.2(d)

```python
""" 2(d) Putting it together to get M 
TODO: calculate M using the functions implemented above """
gamma = 8
### Your code here ###

adjacency_matrix = get_adjacency_matrix(gamma, X)
degree_matrix = get_degree_matrix(adjacency_matrix)
M = np.dot(np.dot(np.diag(1 / np.sqrt(np.diag(degree_matrix))), adjacency_matrix), np.diag(1 / np.sqrt(np.diag(degree_matrix))))


### M = ?
print(M)
```

# Part 1.2(e)

```python
""" 2(e) SVD and Clustering Again 
TODO: run the provided code to perform SVD """
U, S, Vt = np.linalg.svd(M)
top_three_svectors = U[:, :3]

### Your code here: run K-means on top_three_svectors and plot result """
y_pred = KMeans(n_clusters=3, random_state=seed, init = 'random').fit_predict(top_three_svectors)
show_data_results(X, num_plot=2, y_pred=y_pred)
    
""" 2(f) Run the code below to check the actual data """
#show_data_results(X, num_plot=1)
```

Yes, there is much better performance for clustering.  This now matches the optimal clustering expected via data visualization.
