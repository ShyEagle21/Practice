```python
import SC_Simulation_Main_NoVAR as scsm
import SC_Simulation_Main_VAR as scsmv
import pandas as pd


feature_values = {
        'yesterday_total_packages': 8500,
        'RAFT_known_shipped_pkg_count': 2500,
        'RAFT_predicted_carryover_pkg_count': 28,
        'RAFT_predicted_total_handoff_pkg_count': 3500,
        'Day of Week_Sunday': 0,
        'Day of Week_Monday': 1,
        'Day of Week_Tuesday': 0,
        'Day of Week_Wednesday': 0,
        'Day of Week_Thursday': 0,
        'Day of Week_Friday': 0,
        'Day of Week_Saturday': 0,
        'Promotion': 0,
        'TMAX': 36.2,
        'TMIN': 31.8,
        'AWND': 9.87,
        'PRCP': 0.912,
        'SNOW': 6.14
    }

### Labor Allocations ###
night_total_tm = 25

night_tm_pit_unload = 3
night_tm_pit_induct = 6 # must be fewer than 8 or whatever the number of induct spaces is
night_tm_nonpit_split = 1  # must be 1
night_tm_nonpit_NC = 4
night_tm_nonpit_buffer = 11

#NIGHTS TLMD SORT ALLOCATION
night_tm_TLMD_induct = 3  #must be 6 or fewer
night_tm_TLMD_induct_stage = 1
night_tm_TLMD_picker = 4  #Must be 4
night_tm_TLMD_sort = 17

#NIGHTS TLMD STAGE ALLOCATION
night_tm_TLMD_stage = 25


day_total_tm = 15

#DAYS INBOUND INDUCT ALLOCATION
day_tm_pit_unload = 1 
day_tm_pit_induct = 3 # must be fewer than 8 or whatever the number of induct spaces is
day_tm_nonpit_split = 1   # must be 1
day_tm_nonpit_NC = 3
day_tm_nonpit_buffer = 7

#NIGHTS TLMD SORT ALLOCATION
day_tm_TLMD_induct = 2 # must be 6 or fewer
day_tm_TLMD_induct_stage = 1
day_tm_TLMD_picker = 4  #Must be 4
day_tm_TLMD_sort = 8


#DAYSS TLMD STAGE ALLOCATION
day_tm_TLMD_stage = 15


USPS_Fluid_Status = False
UPSN_Fluid_Status = False
FDEG_Fluid_Status = False
FDE_Fluid_Status = False

results_tracker = {}

for i in range(0, 50):
    results = scsm.Simulation_Machine(feature_values,
                        night_total_tm,
                        day_total_tm,
                        night_tm_pit_unload, 
                            night_tm_pit_induct, 
                            night_tm_nonpit_split, 
                            night_tm_nonpit_NC, 
                            night_tm_nonpit_buffer,
                            night_tm_TLMD_induct,
                            night_tm_TLMD_induct_stage,
                            night_tm_TLMD_picker,
                            night_tm_TLMD_sort, 
                            night_tm_TLMD_stage,
                            day_tm_pit_unload,
                            day_tm_pit_induct,
                            day_tm_nonpit_split,
                            day_tm_nonpit_NC,
                            day_tm_nonpit_buffer,
                            day_tm_TLMD_induct,
                            day_tm_TLMD_induct_stage,
                            day_tm_TLMD_picker,
                            day_tm_TLMD_sort,
                            day_tm_TLMD_stage,
                            USPS_Fluid_Status,
                            UPSN_Fluid_Status,
                            FDEG_Fluid_Status,
                            FDE_Fluid_Status)
    results_tracker[i] = results
df = pd.DataFrame.from_dict(results_tracker, orient='index')

# Display the DataFrame
df.to_csv('results.csv')

results_var_tracker = {}

for i in range(0, 50):
    results = scsmv.Simulation_Machine(feature_values,
                        night_total_tm,
                        day_total_tm,
                        night_tm_pit_unload, 
                            night_tm_pit_induct, 
                            night_tm_nonpit_split, 
                            night_tm_nonpit_NC, 
                            night_tm_nonpit_buffer,
                            night_tm_TLMD_induct,
                            night_tm_TLMD_induct_stage,
                            night_tm_TLMD_picker,
                            night_tm_TLMD_sort, 
                            night_tm_TLMD_stage,
                            day_tm_pit_unload,
                            day_tm_pit_induct,
                            day_tm_nonpit_split,
                            day_tm_nonpit_NC,
                            day_tm_nonpit_buffer,
                            day_tm_TLMD_induct,
                            day_tm_TLMD_induct_stage,
                            day_tm_TLMD_picker,
                            day_tm_TLMD_sort,
                            day_tm_TLMD_stage,
                            USPS_Fluid_Status,
                            UPSN_Fluid_Status,
                            FDEG_Fluid_Status,
                            FDE_Fluid_Status)
    results_var_tracker[i] = results
df_var = pd.DataFrame.from_dict(results_var_tracker, orient='index')

# Display the DataFrame
df_var.to_csv('results_var.csv')
```
